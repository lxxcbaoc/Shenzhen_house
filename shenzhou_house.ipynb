```python
#（一）导入数据 合并数据
import os
import pandas as pd

pd.set_option('display.max_columns',None)

data_path=r'D:\BaiduNetdiskDownload\shenzhen_house\data'
save_path=r'D:\BaiduNetdiskDownload\shenzhen_house'

os.chdir(data_path)
loadfile = os.listdir()
df=pd.DataFrame()

for i in loadfile:
    locals()[i.split('.')[0]] =pd.read_excel(i)
    df=pd.concat([df,locals()[i.split('.')[0]]])
df=df.rename(columns={'Unnamed: 0':'houseid'})
print('深圳各区二手房房价信息合并完成')
print('='*80)
```

```python
#(一)检验变量- 每个解释变量是否和被解释变量独立
#1.进行抽样
import pandas as pd
from sklearn.model_selection import StratifiedShuffleSplit

def get_sample(df, sampling='stratified', k=250, stratified_col=None):
    """
    对数据进行抽样。
    
    参数:
    - df: pandas DataFrame，原始数据。
    - sampling: str，抽样方法，默认为'stratified'，即分层抽样。
    - k: int，抽样数量，默认为250。
    - stratified_col: list，分层列名列表，默认为None。
    
    返回:
    - sample_df: pandas DataFrame，抽样后的DataFrame。
    """
    if stratified_col is None:
        raise ValueError("分层列名不能为空")
    
    # 确保分层列是正确的一维数组或列名
    if isinstance(stratified_col, list):
        if len(stratified_col) != 1:
            raise ValueError("分层列名必须是一个元素的列表")
        stratified_col = stratified_col[0]
    
    if stratified_col not in df.columns:
        raise ValueError(f"分层列名 {stratified_col} 不在 DataFrame 中")
    
    # 创建分层抽样对象
    sss = StratifiedShuffleSplit(n_splits=1, test_size=len(df)-k, random_state=0)
    
    # 进行分层抽样
    for train_index, test_index in sss.split(df, df[stratified_col]):
        sample_df = df.iloc[test_index]
    
    return sample_df

data = get_sample(df, sampling='stratified', k=250, stratified_col=['district'])
```

```python
#2.检验各个变量的解释力度
import statsmodels.api as sm
from statsmodels.formula.api import ols
l=['district','roomnum','hall','school','subway']
for i in l:
    model = ols(f'per_price ~ C({i})', data=data).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    p_value = anova_table['PR(>F)'][0]
    print(f'{i}的P值为: {p_value:.4f}')
```

```python
# 对于hall列，由之前的柱状图可知，厅数为3的要明显和其他数量的有较大的区别，对此将其转换头
data['hall_judge']= data['hall'].apply(lambda x:'厅数为3'if x==3 else'厅数不为3')
print(data.head())
```

```python
#对于对分类变量，生成哑变量
data1 = pd.get_dummies(data[['district','C_floor']])# get_dummies特征提取one-ho
print(data1.head())

data1.drop(['district_大鹏新区','C_floor_high'],axis=1,inplace=True)
```

```python
#生成的哑变量与其他所需变量合并成新的数据表
data2 = pd.concat([data1, data[['school', 'subway', 'hall_judge', 'roomnum', 'AREA','hall','per_price']]], axis=1)
print(data2.head())           
```

```python
#1.线性回归模型
import statsmodels.formula.api as smf

lm1 = smf.ols('per_price ~ district_罗湖 + district_坪山 + district_南山 + district_光明 + district_福田 + district_龙华 + district_宝安 + district_盐田 + district_龙岗 ', data=data2).fit()  # 线性回归模型
lm1_summary = lm1.summary()  # 回归结果展示
print(lm1_summary)  # 打印模型摘要
```

```python
data1['predict1']= lm1.predict(data2)
data1['resid1']= lm1.resid
data1.plot('predict1','resid1',kind='scatter',color='#00686b')
plt.show()
```

```python
#2.对数线性模型
data2['per_price_ln']= np.log(data2['per_price'])
data2['AREA_ln']= np.log(data2['AREA'])
#1)只对y取对数
lm2 = smf.ols('per_price_ln ~ district_罗湖 + district_坪山 + district_南山 + district_光明 + district_福田 + district_龙华 + district_宝安 + district_盐田 + district_龙岗 ', data=data2).fit()  # 线性回归模型
lm2_summary= lm2.summary()
print(lm2_summary)

data2['predict2']=lm2.predict(data2)
data2['resid2']= lm2.resid
data2.plot('predict2','resid2',kind='scatter',color='#00686b')
plt.show()
```

```python
#2)对y和x都取对数
import numpy as np
lm3 = smf.ols(
    'per_price_ln ~ np.log(district_罗湖+1) + np.log(district_坪山+1) + np.log(district_南山+1) + np.log(district_光明+1) + np.log(district_福田+1) + np.log(district_龙华+1) + np.log(district_宝安+1) + np.log(district_盐田+1) + np.log(district_龙岗+1)', 
    data=data2).fit()
lm3_summary =lm3.summary()
print(lm3_summary)

data2['predict3']= lm3.predict(data2)
data2['resid3']= lm3.resid
data2.plot('predict3','resid3',kind='scatter',color='#00686b')
plt.show()
```
